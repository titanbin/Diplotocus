# diplotocus.errorbar

````python
import diplotocus as dpl
import matplotlib.pyplot as plt
import numpy as np
````

## diplotocus.errorbar(x, y, xerr=None, yerr=None, duration, delay=0)

[matplotlib API reference.](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html)

````python
from IPython.display import Markdown, display
doc = plt.errorbar.__doc__
display(Markdown(f"```\n{doc}\n```"))
````

### spawn():

````python
x = np.linspace(0,6*np.pi,50)
y = np.cos(x)
xerr = np.random.uniform(0,.1,50)
yerr = np.random.uniform(0,.1,50)

seq = dpl.Sequence()
seq.animate(dpl.errorbar(x=x,y=y,xerr=xerr,yerr=yerr,duration=100).spawn())
seq.save_video(path='../_static/errorbar_spawn.mp4')
````

### morph():

````python
theta = np.linspace(0,2*np.pi,50)
r = 1 - 0.5*np.sin(theta*2)**2
x = np.cos(theta)*r
y = np.sin(theta)*r
xerr = np.random.uniform(0,.1,50)
yerr = np.random.uniform(0,.1,50)

r = 1 + 0.5*np.sin(theta*2)**2
new_x = np.cos(theta)*r
new_y = np.sin(theta)*r
new_xerr = np.random.uniform(0,.1,50)
new_yerr = np.random.uniform(0,.1,50)

seq = dpl.Sequence()
seq.animate(dpl.errorbar(x=x,y=y,xerr=xerr,yerr=yerr,duration=100).morph(new_x=new_x,new_y=new_y,new_xerr=new_xerr,new_yerr=new_yerr))
seq.save_video(path='../_static/errorbar_morph.mp4')
````

### demorph():

```{note}
The `demorph` function should only be called after calling `morph`.
```

````python
theta = np.linspace(0,2*np.pi,50)
r = 1 - 0.5*np.sin(theta*2)**2
x = np.cos(theta)*r
y = np.sin(theta)*r
xerr = np.random.uniform(0,.1,50)
yerr = np.random.uniform(0,.1,50)

r = 1 + 0.5*np.sin(theta*2)**2
new_x = np.cos(theta)*r
new_y = np.sin(theta)*r
new_xerr = np.random.uniform(0,.1,50)
new_yerr = np.random.uniform(0,.1,50)

seq = dpl.Sequence()
errorbar = dpl.errorbar(x=x,y=y,xerr=xerr,yerr=yerr,duration=100).morph(new_x=new_x,new_y=new_y,new_xerr=new_xerr,new_yerr=new_yerr)
seq.animate(errorbar.demorph())
seq.save_video(path='../_static/errorbar_demorph.mp4')
````
