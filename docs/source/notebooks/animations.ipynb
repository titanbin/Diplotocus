# Animations

This page lists the different animations that can be applied to plot objects. All animations share the following parameters:
- `duration`: the number of frames the animation runs from.
- `delay`(optional): the number of frames after what the animation starts playing.
- `easing`(optional): the easing used for this animation [See Easings](easings). If `None`, a linear easing is applied.

## Animations of plot objects

### Translate

`translate(start_pos,end_pos,duration,delay=0,easing=None)` is used to move a plot object from one position to another.
- `start_pos`: where the translation starts, relative to the current center of the plot object.
- `end_pos`: where the translation ends, relative to the current center of the plot object.

```python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

seq = dpl.Sequence(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.translate(start_pos=(0,0),end_pos=(1,1),duration=50)

seq.animate(p)
seq.save_video('../_static/anim_translate.mp4')
```

### Rotate

`rotate(start_angle,end_angle,duration,delay=0,easing=None)` is used to rotate a plot object from one angle to another.
- `start_angle`: the starting angle of the rotation in radians.
- `end_angle`: the ending angle of the rotation in radians.

```python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

seq = dpl.Sequence(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.rotate(start_angle=20,end_angle=110,duration=50)

seq.animate(p)
seq.save_video('../_static/anim_rotate.mp4')
```

### Scale

`scale(start_scale,end_scale,duration,delay=0,easing=None)` is used to scale a plot object from one size to another.
- `start_scale`: the starting scale of the plot object.
- `end_scale`: the ending scale of the plot object.

You can either pass scalar values, in which case the plot object will be uniformly scaled, or a 2-tuple for independent X and Y scales.

```python
import diplotocus as dpl

x = [-1,1,0,-1]
y = [-1,-1,1,-1]

seq = dpl.Sequence(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.scale(start_scale=(0,1),end_scale=(2,0),duration=50)

seq.animate(p)
seq.save_video('../_static/anim_scale.mp4')
```

### Tween

`tween(property,start,end,duration,delay=0,easing=None)` is used to animate a change in a property of the plot object.
- `property`: the name of the property to animate.
- `start`: the starting value of the property.
- `end`: the ending value of the property.

`tweens(properties,starts,ends,duration,delay=0,easing=None)` is used to animate multiple properties at once.
- `properties`: a list of the names of the properties to animate.
- `starts`: the starting values of the properties.
- `ends`: the ending values of the properties.

```python
import diplotocus as dpl

x = [-1,1,0,-1]
y = [-1,-1,1,-1]

seq = dpl.Sequence(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.tween(property='c',start='red',end='green',duration=50)

seq.animate(p)
seq.save_video('../_static/anim_tween.mp4')
```

### Morph

`morph(new_x,new_y,duration,delay=0,easing=None)` is used to morph between the base dataset to a new dataset.
- `new_x`: the x data points to morph to.
- `new_y`: the y data points to morph to.

For plot objects that take 1D datasets (e.g. hist()), `morph()` only accepts `new_x`.

```python
import diplotocus as dpl
import numpy as np

x = np.linspace(0,2*np.pi,100)
y = np.cos(x)

seq = dpl.Sequence(xlim=(-1,7),ylim=(-2,2))

p = dpl.plot(x=x,y=y)

x2 = (np.cos(x)+1)*2
y2 = np.sin(x)
p.morph(new_x=x2,new_y=y2,duration=50)

seq.animate(p)
seq.save_video('../_static/anim_morph.mp4')
```
