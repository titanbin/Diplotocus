# Animations

This page lists the different animations that can be applied to plot objects. All animations share the following parameters:
- `duration`: the number of frames the animation runs from.
- `delay`(optional): the number of frames after what the animation starts playing.
- `easing`(optional): the easing used for this animation [See Easings](easings). If `None`, a linear easing is applied.

Once animations have been set on plot objects, they have to be rendered via `Timeline.animate(plot_objs)` [See Timeline](timeline).

## Animations of plot objects

### Translate

`translate(start_pos,end_pos,duration,delay=0,easing=None)` is used to move a plot object from one position to another.
- `start_pos`: where the translation starts, relative to the current center of the plot object.
- `end_pos`: where the translation ends, relative to the current center of the plot object.

````python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.translate(start_pos=(0,0),end_pos=(1,1),duration=60)

tl.animate(p)
tl.save_video('../_static/anim_translate.mp4')
````

### Rotate

`rotate(start_angle,end_angle,duration,delay=0,easing=None)` is used to rotate a plot object from one angle to another.
- `start_angle`: the starting angle of the rotation in radians.
- `end_angle`: the ending angle of the rotation in radians.

````python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.rotate(start_angle=20,end_angle=110,duration=60)

tl.animate(p)
tl.save_video('../_static/anim_rotate.mp4')
````

### Scale

`scale(start_scale,end_scale,duration,delay=0,easing=None)` is used to scale a plot object from one size to another.
- `start_scale`: the starting scale of the plot object.
- `end_scale`: the ending scale of the plot object.

You can either pass scalar values, in which case the plot object will be uniformly scaled, or a 2-tuple for independent X and Y scales.

````python
import diplotocus as dpl

x = [-1,1,0,-1]
y = [-1,-1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.scale(start_scale=(0,1),end_scale=(2,0),duration=60)

tl.animate(p)
tl.save_video('../_static/anim_scale.mp4')
````

### Tween

`tween(property,start,end,duration,delay=0,easing=None)` is used to animate a change in a property of the plot object.
- `property`: the name of the property to animate.
- `start`: the starting value of the property.
- `end`: the ending value of the property.

`tweens(properties,starts,ends,duration,delay=0,easing=None)` is used to animate multiple properties at once.
- `properties`: a list of the names of the properties to animate.
- `starts`: the starting values of the properties.
- `ends`: the ending values of the properties.


```{note}

`show()` is a shorthand animation that fades in plot objects. It is equivalent to `tween('alpha',0,1)`.

`hide()` is a shorthand animation that fades out plot objects. It is equivalent to `tween('alpha',1,0)`.
```

````python
import diplotocus as dpl

x = [-1,1,0,-1]
y = [-1,-1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.tween(property='c',start='red',end='green',duration=60)

tl.animate(p)
tl.save_video('../_static/anim_tween.mp4')
````

### Morph

`morph(new_x,new_y,duration,delay=0,easing=None)` is used to morph between the base dataset to a new dataset.
- `new_x`: the x data points to morph to.
- `new_y`: the y data points to morph to.

For plot objects that take 1D datasets (e.g. hist()), `morph()` only accepts `new_x`.

````python
import diplotocus as dpl
import numpy as np

x = np.linspace(0,2*np.pi,60)
y = np.cos(x)

tl = dpl.Timeline(xlim=(-1,7),ylim=(-2,2))

p = dpl.plot(x=x,y=y)

x2 = (np.cos(x)+1)*2
y2 = np.sin(x)
p.morph(new_x=x2,new_y=y2,duration=60)

tl.animate(p)
tl.save_video('../_static/anim_morph.mp4')
````

### Draw

`draw(duration,delay=0,easing=None)` is used to animate the plot object by sequentially adding more points of the dataset.
- `reverse`(optional, default `False`): if True, removes points sequentially instead of adding points.

````python
import diplotocus as dpl
import numpy as np

x = np.linspace(0,2*np.pi,60)
y = np.cos(x)

tl = dpl.Timeline(xlim=(-1,7),ylim=(-2,2))

p = dpl.plot(x=x,y=y)
p.draw(duration=60)

tl.animate(p)
tl.save_video('../_static/anim_draw.mp4')
````

### Sequence

`sequence(duration,delay=0,easing=None)` is used to plot one datapoint/row of the dataset per frame. If the dataset is 1-dimensional, this plots a single datapoint per frame, if 2D, it will plot one row per frame. This animation can be used if you have a predefined list of datapoints you want to animate frame-by-frame.

````python
import diplotocus as dpl
import numpy as np

x = np.linspace(0, 2*np.pi, 60)
x_rev = x[::-1]
xs = np.column_stack((x, x_rev))
ys = np.column_stack((np.cos(x), np.sin(x_rev)))

tl = dpl.Timeline(xlim=(-1,7),ylim=(-2,2))

p = dpl.plot(x=xs,y=ys,lw=0,marker='+',ms=10)
p.sequence(duration=50)

tl.animate(p)
tl.save_video('../_static/anim_sequence.mp4')
````

## Animations of figures and axes

### Axis zoom

`axis_zoom(zoom,duration,delay=0,easing=None,axis=None)` is used to zoom within an axis by changing its limits while keeping the center constant.
- `zoom`: the zoom value to end on (multiplicative factor of the current axis ranges).
- `axis`(optional): the axis to zoom in, if multiple axes exist. If `None`, will act on the main axis of the Timeline.

````python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))
tl.main_axis.plot(x,y)

a = dpl.axis_zoom(zoom=2,duration=60)

tl.animate(a)
tl.save_video('../_static/axis_zoom.mp4')
````

### Axis limits

`axis_limits(duration,xlim=None,ylim=None,delay=0,easing=None,axis=None)` is used to change the limits of an axis to a given range.
- `xlim`(optional): the new limits of the X-axis, if left empty then left unchanged.
- `ylim`(optional): the new limits of the Y-axis, if left empty then left unchanged.
- `axis`(optional): the axis to zoom in, if multiple axes exist. If `None`, will act on the main axis of the Timeline.

````python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))
tl.main_axis.plot(x,y)

a = dpl.axis_limits(xlim=(-5,5),ylim=(-1,1),duration=60)

tl.animate(a)
tl.save_video('../_static/axis_limits.mp4')
````

### Axis move

`axis_move(end_pos,duration,start_pos=None,delay=0,easing=None,axis=None)` is used to move the center of an axis from and to given values.
- `end_pos`: the new limits of the Y-axis, if left empty then left unchanged.
- `start_pos`(optional): the initial axis center. If `None`, the current center is taken.
- `axis`(optional): the axis to zoom in, if multiple axes exist. If `None`, will act on the main axis of the Timeline.

````python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))
tl.main_axis.plot(x,y)

a = dpl.axis_move(end_pos=(1,1),duration=60)

tl.animate(a)
tl.save_video('../_static/axis_move.mp4')
````

### Axis alpha

`axis_alpha(start_alpha,end_alpha,duration,alpha_objs=True,delay=0,easing=None,axis=None)` is used to animate the opacity of an axis.
- `start_alpha`: the initial alpha value.
- `end_alpha`: the final alpha value.
- `alpha_objs`(optional, default `True`): whether to apply the alpha to plotted objects or just the axis.
- `axis`(optional): the axis to zoom in, if multiple axes exist. If `None`, will act on the main axis of the Timeline.

````python
import diplotocus as dpl

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

tl = dpl.Timeline(xlim=(-2.5,2.5),ylim=(-2,2))
tl.main_axis.plot(x,y)

a = dpl.axis_alpha(start_alpha=0,end_alpha=1,duration=50)

tl.animate(a)
tl.save_video('../_static/axis_alpha.mp4')
````

### Figure width ratio

`fig_width_ratio(start_widths,end_widths,duration,delay=0,easing=None)` is used to animate the width ratios of figures with mutiple columns of subplots.
- `start_widths`: the initial width ratios (for a figure with n columns, a n-long iterator).
- `end_widths`: the final width ratios (for a figure with n columns, a n-long iterator).

````python
import diplotocus as dpl
import matplotlib.pyplot as plt

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x,y)
ax[1].plot(x,y)
tl = dpl.Timeline(fig=fig)

a = dpl.fig_width_ratio(start_widths=(1,0),end_widths=(0,1),duration=60)

tl.animate(a)
tl.save_video('../_static/fig_width_ratio.mp4')
````

### Figure height ratio

`fig_height_ratio(start_heights,end_heights,duration,delay=0,easing=None)` is used to animate the heights ratios of figures with mutiple rows of subplots.
- `start_heights`: the initial height ratios (for a figure with m rows, a m-long iterator).
- `end_heights`: the final height ratios (for a figure with m rows, a m-long iterator).

````python
import diplotocus as dpl
import matplotlib.pyplot as plt

x = [-1,1,1,-1,-1]
y = [-1,-1,1,1,-1]

fig,ax = plt.subplots(2,1,figsize=(7,7))
ax[0].plot(x,y)
ax[1].plot(x,y)
tl = dpl.Timeline(fig=fig)

a = dpl.fig_height_ratio(start_heights=(1,0),end_heights=(0,1),duration=60)

tl.animate(a)
tl.save_video('../_static/fig_height_ratio.mp4')
````
