# Plotting functions

This page lists all implemented plotting functions from matplotlib.

```python
import diplotocus as dpl
import matplotlib.pyplot as plt
import numpy as np
```

## diplotocus.plot(x, y, duration, delay=0):

```python
x = np.linspace(0,6*np.pi,200)
y = np.cos(x)

seq = dpl.Sequence()
seq.animate(dpl.plot(x=x,y=y,duration=100).spawn())
seq.save_video(path='_static/plot.mp4')
```

## diplotocus.scatter(x, y, duration, delay=0):

```python
x = np.linspace(0,6*np.pi,200)
y = np.cos(x)

seq = dpl.Sequence()
seq.animate(dpl.scatter(x=x,y=y,duration=100).spawn())
seq.save_video(path='_static/scatter.mp4')
```
