# diplotocus.axvline

````python
import diplotocus as dpl
import matplotlib.pyplot as plt
import numpy as np
````

## diplotocus.axvline(x, duration, delay=0)

[matplotlib API reference.](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axvline.html)

````python
from IPython.display import Markdown, display
doc = plt.axvline.__doc__
display(Markdown(f"```\n{doc}\n```"))
````

### spawn():

````python
x = 0.5

seq = dpl.Sequence()
seq.animate(dpl.axvline(x=x,duration=100).spawn())
seq.save_video(path='../_static/axvline_spawn.mp4')
````

### despawn():

````python
x = 0.5

seq = dpl.Sequence()
seq.animate(dpl.axvline(x=x,duration=100).despawn())
seq.save_video(path='../_static/axvline_despawn.mp4')
````

### show():

````python
x = 0.5

seq = dpl.Sequence()
seq.animate(dpl.axvline(x=x,duration=100).show())
seq.save_video(path='../_static/axvline_show.mp4')
````

### hide():

````python
x = 0.5

seq = dpl.Sequence()
seq.animate(dpl.axvline(x=x,duration=100).hide())
seq.save_video(path='../_static/axvline_hide.mp4')
````
